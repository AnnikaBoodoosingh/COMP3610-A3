```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# Collect counts from all files
category_counter = Counter()

for path in parquet_paths:
    try:
        df = pd.read_parquet(path, engine='pyarrow', columns=['main_category'])  # only load the needed column
        category_counter.update(df['main_category'].dropna())
    except Exception as e:
        print(f"Error reading {path}: {e}")

# Get top 10 categories by count
top_10 = category_counter.most_common(10)
categories, counts = zip(*top_10)

# Plot
plt.figure(figsize=(12, 6))
plt.bar(categories, counts, color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Categories by Review Count (main_category)')
plt.ylabel('Number of Reviews')
plt.tight_layout()
plt.show()
```
