```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import dask.dataframe as dd
```

```python
# Load Parquet files with Dask
df = dd.read_parquet(parquet_paths, columns=['rating'])

# Convert the ratings to a numpy array and drop null values
ratings = df['rating'].dropna().compute()  # Compute returns the actual data in memory
```

```python
# Now plot the histogram for ratings 1 to 5
plt.figure(figsize=(8, 5))
plt.hist(ratings, bins=[0.5, 1.5, 2.5, 3.5, 4.5, 5.5], edgecolor='black', color='skyblue', kde = True)

# Add labels and title
plt.title('Star Rating Histogram (1–5)')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.xticks([1, 2, 3, 4, 5])  # Set x-axis labels to be 1-5
plt.grid(axis='y')

# Show the plot
plt.show()
```

```python
import seaborn as sns
```

```python
# 3) Plot histogram with Seaborn
plt.figure(figsize=(8, 5))
sns.histplot(
    data=ratings,
    bins=[0.5,1.5,2.5,3.5,4.5,5.5],  # center bins on each integer rating
    discrete=True,
    kde=True
)
plt.xlabel('Star Rating')
plt.ylabel('Count of Reviews')
plt.title('Histogram of Star Ratings (1–5)')
plt.tight_layout()
plt.show()
```
